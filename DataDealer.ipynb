```python
# 处理turnover=0的情况，利用市场和历史平均换手加权填充
t = daily.turnover_ratio.T
market = t.where(t!=0).mean(axis=1)
history = t.where(t!=0).mean(axis=0)

market = pd.DataFrame({stock:market for stock in t.columns})
history = pd.DataFrame({date:history for date in t.index}).T

q = 0.6
fillval = market*q+(1-q)*history
t = t.where(t!=0, fillval)
daily.turnover_ratio = t.T
```

```python
daily_data = pd.read_csv('arranged_data_20140102_20211029.csv').set_index(['trade_date', 'stock_code']).sort_index()
daily_data['soe'] = np.zeros((daily_data.shape[0],))
issoe = pd.read_csv('issoe.csv').values
```

```python
def updatesoe(x):
    if x.name in issoe:
        x.soe = 1
    return x

daily_data = daily_data.groupby('stock_code').apply(updatesoe)
```
